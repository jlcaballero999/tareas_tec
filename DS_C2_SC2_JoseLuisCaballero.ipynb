**Aplicación de todas las vacunas en el mundo con datos Our World in Data **


1.	Extraer la información del archivo.

```python
import pandas as pd
dbcovid= pd.read_csv('/01_COVID/country_vaccinations.csv')
dbcovid.head()


```

2.	Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
####Cambiar tipo tiempo ----
dbcovid['date'] = pd.to_datetime(dbcovid['date'])
####tipo de dato y estrctura ----
dbcovid.info()
```

3.	Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
### agrupar por vaccines y realice la sumatoria----
vaccines_summary = dbcovid.groupby('vaccines')['total_vaccinations'].sum().reset_index()
vaccines_summary
```

4.	Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
#### cantidad de vacunas aplicadas en el mundo----
vacunas_mundo = dbcovid['total_vaccinations'].sum()
vacunas_mundo
```

5.	Calcular el promedio de vacunas aplicadas por país.

```python
media_vac_por_pais = dbcovid.groupby('country')['total_vaccinations'].mean().reset_index()
media_vac_por_pais.head(8)
```

6.	Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
vacunas_29_01_21 = dbcovid[dbcovid['date'] == '2021-01-29']['total_vaccinations'].sum()
vacunas_29_01_21
```

7.	Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
conDiferencias=dbcovid.assign(diferencias=dbcovid['daily_vaccinations'] - dbcovid['daily_vaccinations_raw'])
conDiferencias.head()
```

8.	Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
fecha_mas_reciente = dbcovid['date'].max()
fecha_mas_antigua = dbcovid['date'].min()
periodo_tiempo = fecha_mas_reciente - fecha_mas_antigua
print("Fecha más reciente:", fecha_mas_reciente)
print("Fecha más antigua:", fecha_mas_antigua)
print("Periodo de tiempo:", periodo_tiempo)

```

9.	Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
conCantidad = dbcovid.assign(canVac=dbcovid['vaccines'].str.split(',').str.len())
conCantidad.head()
```

10.	Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20=dbcovid[dbcovid['date'] < '2020-12-20']
antes20.head()
#ultima_fecha = antes20['date'].max()
#print("fecha más reciente:", ultima_fecha)
```

11.	Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
# todo en donde la columna vaccines aparezca la palabra Pfizer
pfizer = dbcovid[dbcovid['vaccines'].str.contains('Pfizer', na=False)]
pfizer.head()
pfizer.info()

```

12.	Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.
